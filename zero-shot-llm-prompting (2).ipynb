```python
import pandas as pd
import re

# Load the dataset
df = pd.read_excel("/kaggle/input/validationdataset/Final Data File_Validation.xlsx")
df = df.drop(index=0).reset_index(drop=True)

# Print column names to help debug
print("Available columns in the DataFrame:")
for col in df.columns:
    print(f"- {col}")

def generate_prompt(row):
    """Generate an evaluation prompt for a specific row of survey data."""
    
    # Function to safely get values from DataFrame
    def safe_get(row, column_name, partial_match=False):
        if column_name in row:
            return row[column_name]
        
        if partial_match:
            # Try to find a column that contains the key text
            matching_cols = [col for col in row.index if column_name in col]
            if matching_cols:
                return row[matching_cols[0]]
        
        # Return a default value if column not found
        return "Not Available"
    
    # Identify consumed alcoholic beverages (Q6)
    # Using partial column name matching
    q6_prefix = "Q6 Which of the following types of alcoholic beverages"
    
    alcohol_columns = {
        "Beer": [col for col in df.columns if q6_prefix in col and "Alcohol Category" in col][0] if any(q6_prefix in col for col in df.columns) else "Not Found",
        "Flavored Mixed Beer": "Unnamed: 8",
        "Non-Alcoholic Beers": "Unnamed: 9",
        "Hard Ciders": "Unnamed: 10",
        "Hard Kombucha": "Unnamed: 11",
        "Wine": "Unnamed: 12",
        "Hard Beverage": "Unnamed: 13",
        "Distilled Spirits": "Unnamed: 14"
    }
    
    consumed_alcohol = []
    for name, col in alcohol_columns.items():
        try:
            if col != "Not Found" and row[col] == "Yes":
                consumed_alcohol.append(name)
        except KeyError:
            print(f"Warning: Column '{col}' not found for '{name}'")
    
    consumed_alcohol_str = ", ".join(consumed_alcohol) if consumed_alcohol else "None"

    # Identify consumed beer categories (Q7)
    q7_prefix = "Q7. Which of the following beer types"
    
    beer_columns = {
        "Domestic Light Beer": [col for col in df.columns if q7_prefix in col and "Beer Category" in col][0] if any(q7_prefix in col for col in df.columns) else "Not Found",
        "Domestic Regular Beer": "Unnamed: 16",
        "Low-Calorie/Carb Beer": "Unnamed: 17",
        "Domestic Craft": "Unnamed: 18",
        "Imported Mexican Beer": "Unnamed: 19",
        "Other Imported Beer": "Unnamed: 20",
        "Non-Alcoholic Beer": "Unnamed: 21"
    }

    consumed_beers = []
    for name, col in beer_columns.items():
        try:
            if col != "Not Found" and row[col] == "Yes":
                consumed_beers.append(name)
        except KeyError:
            print(f"Warning: Column '{col}' not found for '{name}'")
    
    consumed_beers_str = ", ".join(consumed_beers) if consumed_beers else "None"

    # Create survey structure description
    survey_structure = "A concept testing survey evaluating consumer reactions to ULTRA Pure Gold, a new premium light beer positioned as a 'Superior Light Beer' with low carbs (2.6g) and calories (85), among adults aged 21-65 who consume alcoholic beverages."

    # Define column mappings with possible variations
    column_mappings = {
        "gender": [col for col in df.columns if "gender" in col.lower()][0] if any("gender" in col.lower() for col in df.columns) else "Not Found",
        "age": [col for col in df.columns if "age" in col.lower()][0] if any("age" in col.lower() for col in df.columns) else "Not Found",
        "urban_rural": [col for col in df.columns if "area or community" in col.lower()][0] if any("area or community" in col.lower() for col in df.columns) else "Not Found",
        "income": [col for col in df.columns if "household income" in col.lower()][0] if any("household income" in col.lower() for col in df.columns) else "Not Found",
        "relevance": [col for col in df.columns if "relevant" in col.lower() and "concept relevance" in col.lower()][0] if any("relevant" in col.lower() and "concept relevance" in col.lower() for col in df.columns) else "Not Found",
        "appeal": [col for col in df.columns if "appealing" in col.lower() and "concept appeal" in col.lower()][0] if any("appealing" in col.lower() and "concept appeal" in col.lower() for col in df.columns) else "Not Found",
        "differentiation": [col for col in df.columns if "different" in col.lower() and "concept differentiation" in col.lower()][0] if any("different" in col.lower() and "concept differentiation" in col.lower() for col in df.columns) else "Not Found",
        "believability": [col for col in df.columns if "believable" in col.lower() and "concept beleivability" in col.lower()][0] if any("believable" in col.lower() and "concept beleivability" in col.lower() for col in df.columns) else "Not Found",
        "price": [col for col in df.columns if "price" in col.lower() and "concept_price" in col.lower()][0] if any("price" in col.lower() and "concept_price" in col.lower() for col in df.columns) else "Not Found",
        "purchase_intent": [col for col in df.columns if "likely" in col.lower() and "purchase intent" in col.lower()][0] if any("likely" in col.lower() and "purchase intent" in col.lower() for col in df.columns) else "Not Found",
        "drinking_frequency": [col for col in df.columns if "drink" in col.lower() and "frequency" in col.lower()][0] if any("drink" in col.lower() and "frequency" in col.lower() for col in df.columns) else "Not Found",
        "replacement_product": [col for col in df.columns if "replacement product" in col.lower()][0] if any("replacement product" in col.lower() for col in df.columns) else "Not Found",
        "like": [col for col in df.columns if "like about the shown concept" in col.lower()][0] if any("like about the shown concept" in col.lower() for col in df.columns) else "Not Found",
        "dislike": [col for col in df.columns if "dislike about the shown concept" in col.lower()][0] if any("dislike about the shown concept" in col.lower() for col in df.columns) else "Not Found",
    }
    
    # Print column mappings once
    if not hasattr(generate_prompt, "mappings_printed"):
        print("\nColumn mappings found:")
        for key, value in column_mappings.items():
            print(f"- {key}: {value}")
        generate_prompt.mappings_printed = True

    # Create the prompt with safe column access
    prompt = f"""
    You are a strict evaluator for beer survey responses. Analyze the response below and output:
1 - If the response is low quality (Contradiction,verly Positive,Vague Response, Minimal detail in "dislike" response  ,Lack of Depth:* "Like" is clear, "Dislike" is brief  ,Inconsistencies,Low Effort
0 - If the response is good quality (relevant, informative, appropriate)
Survey Structure: {survey_structure}

USER PERSONA  
- {safe_get(row, column_mappings['gender'])}, {safe_get(row, column_mappings['age'])}, {safe_get(row, column_mappings['urban_rural'])}, Income: {safe_get(row, column_mappings['income'])}
- Alcohol Types Consumed over last 4 weeks: {consumed_alcohol_str}
- Beer Types Consumed over last 4 weeks: {consumed_beers_str}

PRODUCT PREFERENCE  
- Relevance: {safe_get(row, column_mappings['relevance'])} & Appeal: {safe_get(row, column_mappings['appeal'])}
- Differentiation: {safe_get(row, column_mappings['differentiation'])} & {safe_get(row, column_mappings['believability'])}
- Price: {safe_get(row, column_mappings['price'])}
- Purchase Intent: {safe_get(row, column_mappings['purchase_intent'])}
- Consumption (if free): {safe_get(row, column_mappings['drinking_frequency'])}
- Replacement Preference: {safe_get(row, column_mappings['replacement_product'])}

USER RESPONSES  
- *Like:* "{safe_get(row, column_mappings['like'])}"
- *Dislike:* "{safe_get(row, column_mappings['dislike'])}"

Evaluation Criteria(strict): 
- *Contradiction:*   
- *Overly Positive:*   
- *Vague Response:* Minimal detail in "dislike" response  
- *Lack of Depth:* "Like" is clear, "Dislike" is brief  
- *Inconsistencies:* None  
- *Low Effort:* "Dislike" response is minimal 

Flag the response ( 1 = Bad Response, 0 = Good response,) and only(either 0 or 1) and nothing else in the output.
Return ONLY the number 0 or 1 with no explanation or additional text.
"""
    return prompt

# Function to extract quality flag from LLM response
def extract_flag(response):
    match = re.search(r'\b[01]\b', response)
    if match:
        return int(match.group())
    else:
        # Default to 0 if no valid flag (0 or 1) is found
        return 0

# Process a batch of rows
def process_batch(start_idx, end_idx, llm, tokenizer, sampling_params):
    results = []
    
    for idx in range(start_idx, min(end_idx, len(df))):
        try:
            row = df.iloc[idx]
            llm_prompt = generate_prompt(row)
            
            # Format the prompt using your specific method
            messages = [
                {
                    "role": "user",
                    "content": llm_prompt
                },
                {"role": "assistant", "content": ""}  # Placeholder
            ]
            
            text = tokenizer.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=True
            )
            
            # Generate outputs using your method
            outputs = llm.generate([text], sampling_params)
            
            # Process the generated text
            for output in outputs:
                generated_text = output.outputs[0].text
                flag = extract_flag(generated_text)
                
                results.append({
                    'row_idx': idx,
                    'flag': flag,
                    'response': generated_text
                })
                
                print(f"Row {idx} - Flag: {flag}")
            
        except Exception as e:
            print(f"Error processing row {idx}: {str(e)}")
            # Default to 0 on error
            results.append({
                'row_idx': idx,
                'flag': 0,
                'response': f"Error: {str(e)}"
            })
    
    return results

# Process a single row
def process_single_row(row_number, llm, tokenizer, sampling_params):
    row = df.iloc[row_number]
    llm_prompt = generate_prompt(row)
    
    # Format the prompt using your specific method
    messages = [
        {
            "role": "user",
            "content": llm_prompt
        },
        {"role": "assistant", "content": ""}  # Placeholder
    ]
    
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    
    # Generate outputs using your method
    outputs = llm.generate([text], sampling_params)
    
    # Process the generated text
    for output in outputs:
        prompt = output.prompt
        generated_text = output.outputs[0].text
        flag = extract_flag(generated_text)
        
        print(f"Row {row_number} - Flag: {flag}")
        print(f"Generated text: {generated_text!r}")
    
    return flag, generated_text

# Process all rows in the dataframe
def process_all_rows(df, llm, tokenizer, sampling_params):
    results = []
    
    for idx in range(len(df)):
        try:
            row = df.iloc[idx]
            llm_prompt = generate_prompt(row)
            
            # Format the prompt using your specific method
            messages = [
                {
                    "role": "user",
                    "content": llm_prompt
                },
                {"role": "assistant", "content": ""}  # Placeholder
            ]
            
            text = tokenizer.apply_chat_template(
                messages,
                tokenize=False,
                add_generation_prompt=True
            )
            
            # Generate outputs using your method
            outputs = llm.generate([text], sampling_params)
            
            # Process the generated text
            for output in outputs:
                generated_text = output.outputs[0].text
                flag = extract_flag(generated_text)
                
                results.append({
                    'row_idx': idx,
                    'flag': flag,
                    'response': generated_text
                })
            
        except Exception as e:
            print(f"Error processing row {idx}: {str(e)}")
            # Default to 0 on error
            results.append({
                'row_idx': idx,
                'flag': 0,
                'response': f"Error: {str(e)}"
            })
        
        # Print progress every 10 rows
        if idx % 10 == 0:
            print(f"Processed {idx} rows")
    
    return results

# Apply results to the dataframe and save
def apply_results_to_df(results):
    for result in results:
        df.at[result['row_idx'], 'OE_Quality_Flag'] = result['flag']
    
    # Save the updated dataframe
    df.to_excel("/kaggle/working/Final_Data_File_Flagged1.xlsx", index=False)
    print("Updated dataframe saved to /kaggle/working/Final_Data_File_Flagged12.xlsx")
    
    return df

# Example usage:
# Process row 196 (as in your example)
# flag, response = process_single_row(196, llm, tokenizer, sampling_params)

# Process a batch of rows (e.g., rows 0-100)
# batch_results = process_batch(0, 100, llm, tokenizer, sampling_params)
# updated_df = apply_results_to_df(batch_results)

# Process all rows
all_results = process_all_rows(df, llm, tokenizer, sampling_params)
final_df = apply_results_to_df(all_results)
```
